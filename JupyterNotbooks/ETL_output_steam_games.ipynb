# Proceso de ETL del Dataset 'output_steam_games.json'

```python
# Importacion de las librerias necesarias
import pandas as pd
import json
%pip install pyarrow
%load_ext autoreload
%autoreload 2
import utils
import warnings
warnings.filterwarnings("ignore")
```

## Extraccion de datos del Dataset

```python
# Leemos el archivo 'output_steam_games.json' y lo convertimos directamente en un DataFrame
with open('../Datasets/output_steam_games.json') as f:
    filas_games = [json.loads(line) for line in f]
df_steam_games  = pd.DataFrame(filas_games)
# Desplegamos el DataFrame
df_steam_games
```

```python
# Eliminar las filas que contienen valores nulos en todas las columnas 
df_steam_games.dropna(how='all', inplace=True)
# Reiniciar el Ã­ndice del DataFram
df_steam_games.reset_index(drop=True, inplace=True)
# Desplegar las dimensiones del dataframe
df_steam_games.shape
```

## Trabajar con las columanas 'publisher', 'app_name', 'title' y 'developer'

```python
# Columnas a rellenar
columnas_incompletas = ['publisher', 'app_name', 'title', 'developer']
# Rellena los nulos en las columnas seleccionadas
df_steam_games[columnas_incompletas] = df_steam_games[columnas_incompletas].fillna('Sin dato disponible')
# Muestra el dataframe resultante
df_steam_games.head()
```

## Trabajar con la columna 'genres'

```python
# Eliminar las filas que contienen valores nulos en la Columna 'genres'
df_steam_games = df_steam_games.explode('genres').dropna(subset=['genres']).head()
# Desplegar el DataFrame
df_steam_games
```

# Trabajar con las columna 'tags', 'specs', 'url', 'reviews_url']

```python
# Eliminar las Columnas 'tags', 'specs', 'url', 'reviews_url' 
df_steam_games.drop(['tags', 'specs', 'url', 'reviews_url'], axis=1, inplace=True)
df_steam_games
```

## Trabajo con la columna 'release_date'

```python
# Convertir la Columna 'release_date' a un formato de fecha y hora
df_steam_games['release_date'] = pd.to_datetime(df_steam_games['release_date'],dayfirst=True, errors='coerce')
```

```python
# Crear la columna 'release_year' a partir de la Columna 'release_date' 
df_steam_games['release_year'] = df_steam_games['release_date'].dt.year
# Eliminar la columna 'release_date' 
df_steam_games.drop('release_date', axis=1, inplace=True)
# Desplegar el DataFrame
df_steam_games.head()
```

## Trabajo con la columna 'price'

```python
# Reemplazar valores string en la columna 'price' por 0
df_steam_games['price'].replace('Free to Play', 0, inplace=True)
df_steam_games['price'].replace('Free To Play', 0, inplace=True)
df_steam_games['price'].replace('Play the Demo', 0, inplace=True)
df_steam_games['price'].replace('Install Now', 0, inplace=True)
df_steam_games['price'].replace('Free Mod', 0, inplace=True)
df_steam_games['price'].replace('Play for Free!', 0, inplace=True)
df_steam_games['price'].replace('Play WARMACHINE: Tactics Demo', 0, inplace=True)
df_steam_games['price'].replace('Install Theme', 0, inplace=True)
df_steam_games['price'].replace('Third-party', 0, inplace=True)
df_steam_games['price'].replace('Play Now', 0, inplace=True)
df_steam_games['price'].replace('Free HITMANâ„¢ Holiday Pack', 0, inplace=True)
df_steam_games['price'].replace('Free Movie', 0, inplace=True)
df_steam_games['price'].replace('Free', 0, inplace=True)
df_steam_games['price'].replace('Free Demo', 0, inplace=True)
df_steam_games['price'].replace('Free to Try', 0, inplace=True)
df_steam_games['price'].replace('Starting at $499.00', 0, inplace=True)
# Desplegar el DataFrame
df_steam_games
```

```python
# Especificar que los valores de la columna 'price' sean tratados como floats 
df_steam_games['price'] = df_steam_games['price'].astype(float)
# Desplegar el tipo de dato de la columna dtype para verificar que no sea de tipo string
df_steam_games['price'].dtype
```

```python
# Eliminar todos los corchetes de la columna 'Genres'
df_steam_games['genres'] = df_steam_games['genres'].str.replace('[', '').str.replace(']', '')
# Eliminar las comillas simples y los corchetes de cierre
df_steam_games['genres'] = df_steam_games['genres'].str.replace("'", '')
# Desplegar DataFrame
df_steam_games
```
